## Additionnal data from CSAbstruct

```python
import pandas as pd
```

```python
train = pd.read_parquet("train-csabstruct.parquet")
eval = pd.read_parquet("validation-csabstruct.parquet")
test = pd.read_parquet("test-csabstruct.parquet")

df_all = pd.concat([train, eval, test])
print(df_all.shape)

df_all.head()
```

```python
label2id = {
    "background":0,
    "method": 1,
    "objective": 2,
    "other": 3,
    "result":4,
}

id2label = {v:k for k,v in label2id.items()}
```

```python
data = []

for i, row in df_all.iterrows():
    
    sentences = row["sentences"]
    labels = row["labels"]
    confs = row["confs"]
    abstract_id = row["abstract_id"]

    for j, sent in enumerate(sentences):
        data.append([abstract_id, sent, id2label[labels[j]], confs[j]])

df = pd.DataFrame(data, columns = ["abstract_id", "text", "label", "confidence_score"])
print(df.shape)
```
